```python
# importing libraries
import spotipy
import time
import time
import numpy as np
import pandas as pd
```

```python
from spotipy.oauth2 import SpotifyClientCredentials
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="client_id",
                                                           client_secret="client_secret"))

```

```python
# a function to get a song list woth their audio feature from spotify and store them in a dataframe
def audio_features_spotify(artist):
    results = sp.search(q=f'artist:{artist}', limit=50)
    # extracting the track ids
    track_ids = [track['id'] for track in results['tracks']['items']]
    song_names = [track['name'] for track in results['tracks']['items']]
    # extracting the audio features
    audio_features = sp.audio_features(track_ids)
    # storing the result in a dataframe
    df = pd.DataFrame(audio_features)
    df['artist'] = artist
    df['song name'] = song_names
    return df
```

```python
# getting list of artist and apply the function to get songs
name = ["Micheal Jackson","Christina","Shakira", "Celine Dion", "Adele","Justin Bieber", "Nirvana", "pink floyd", "Ebi"]
#name = ["Micheal Jackson","Christina","Celine Dion" ]
df = pd.DataFrame()

for artist in name:
    df_artist = audio_features_spotify(artist)
    df = pd.concat([df, df_artist])
    
df = df.reset_index(drop=True)
df
```

```python
#saving as a csv file
df.to_csv("spotify_song_data.csv")
```
